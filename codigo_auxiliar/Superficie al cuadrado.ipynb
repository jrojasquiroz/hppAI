# ¿Por qué incluiríamos una variable que sea la superficie al cuadrado en nuestro modelo?

Porque muy probablemente hay un umbral a partir del cual el tamaño de un piso deja de ser un factor tan decisivo para su precio. Si, en una situación hipotética, a partir de 300 m$^2$ todas las familias tuvieran todas sus necesidades cubiertas, ¿no tendría sentido que el precio deje de aumentar a partir de esa superficie?

Esta variable captura esa complejidad. Noten los gráficos de dispersión.

```python
import pandas as pd #permite manipular los dataframes, que serían el equivalente a los Excel
import numpy as np #en caso se necesiten cálculos numéricos, como p. ej. encontrar valores nulos 

base=pd.read_spss('rawdata/clases/habitaclia4043.csv')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos del DataFrame 'base'
precios = base['precio_eur'].values
superficie = base['superficie'].values
superficie_2 = base['superficie_2'].values

# Creación de subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de dispersión para superficie
axs[0].scatter(superficie, precios, label='Datos', alpha=0.7, s=20)
axs[0].set_xlabel('Superficie (metros cuadrados)')
axs[0].set_ylabel('Precio (EUR)')
axs[0].legend()
axs[0].set_title('Relación entre Precio y Superficie de la Propiedad')

# Gráfico de dispersión para superficie al cuadrado
axs[1].scatter(superficie_2, precios, label='Datos', alpha=0.7, s=20)
axs[1].set_xlabel('Superficie al Cuadrado')
axs[1].set_ylabel('Precio (EUR)')
axs[1].legend()
axs[1].set_title('Relación entre Precio y Superficie al Cuadrado')

plt.tight_layout()
plt.show()
```

A partir de aquí tendrían 3 opciones:
1. Hacer modelos sólo con la superficie, porque considera que la superficie al cuadrado es difícil de interpretar o explicar.
2. Hacer modelos sólo con la superficie al cuadrado, porque consideran que refleja mejor la relación compleja que existe entre la superficie y el precio de los pisos.
3. Hacer un modelo con ambas variables, pero revisando el VIF para evitar problemas (noten que la correlación entre estas variables es bastante fuerte).

```python
from scipy.stats import pearsonr

# Calcular el coeficiente de correlación y el p-valor
correlation_coef, p_value = pearsonr(superficie, superficie_2)

print("Coeficiente de correlación entre superficie y superficie al cuadrado:", correlation_coef)
print("P-valor asociado:", p_value)
```
